## Bar Plot - Categorical Data

```python
# 1. Bar Plot: Distribution of the 'ChestPain' column
#. value_counts() : counts how many times chest pain appears in the dataset
chest_pain_counts = df['ChestPain'].value_counts()
plt.figure(figsize=(6, 4))
#plots the number of times each chest pain (y-axis, .index) occurs vs the values (x-axis, .values)
plt.bar(chest_pain_counts.index, chest_pain_counts.values, color='red')
plt.title('Bar Plot: Distribution of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

## Bivariate Analysis

## Box Plot - Displays the spread and outliers of data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Box Plot: Age vs Chest Pain Type
plt.figure(figsize=(8, 5))
sns.boxplot(x='ChestPain', y='Age', data=df, palette='Set2')
plt.title('Box Plot: Age vs Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Age')
plt.show()
```

## Conclusion: "Typical" and "Asymptomatic" chest pain have a slightly higher median age than "Nonanginal" and "Nontypical."

## Scatter Plot - Correlation between two variables

```python
# 2. Scatter Plot: RestBP vs MaxHR, colored by AHD (Heart Disease)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='RestBP', y='MaxHR', hue='AHD', data=df, palette='viridis')
plt.title('Scatter Plot: Resting BP vs Max Heart Rate')
plt.xlabel('Resting Blood Pressure (RestBP)')
plt.ylabel('Maximum Heart Rate (MaxHR)')
plt.legend(title='Heart Disease')
plt.show()
```

## Conclusion: People with higher blood pressure do not necessarily have a higher max heart rate.Those with heart disease (green dots) tend to have lower MaxHR at the same BP levels, suggesting that heart disease affects how high their heart rate can go.

## Line Plot - showcases trends over time

```python
# 3. Line Plot: Age vs Cholesterol
plt.figure(figsize=(8, 5))
sns.lineplot(x='Age', y='Chol', data=df, color='purple', marker='o')
plt.title('Line Plot: Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

## Conclusion: Cholestrol has more risks after the age of 60.

## Violin Plot - numerical data distribution with density curves

```python
# 4. Violin Plot: MaxHR vs Chest Pain Type
plt.figure(figsize=(8, 5))
sns.violinplot(x='ChestPain', y='MaxHR', data=df, palette='muted')
plt.title('Violin Plot: Max Heart Rate vs Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Maximum Heart Rate')
plt.show()
```

## Conclusion: Typical and Nonanginal chest pain types tend to have higher MaxHR distributions.Asymptomatic patients seem to have lower MaxHR values overall, which might indicate an underlying cardiovascular issue.There is no drastic difference in MaxHR across the chest pain types, but asymptomatic cases appear more at risk of lower MaxHR.

## Multivariate Analysis

## Pair Plot - scatter plot for multiple variables

```python
# 1. Pair Plot: Analyze pairwise relationships between key numerical variables
plt.figure(figsize=(10, 8))
sns.pairplot(df, vars=['Age', 'RestBP', 'Chol', 'MaxHR', 'Oldpeak'], hue='AHD', palette='Set2')
plt.suptitle('Pair Plot: Multivariate Analysis', y=1.02)
plt.show()
```

## Conclusion: However, we see that people with lower MaxHR (more orange dots in the lower MaxHR range) tend to have higher chances of heart disease.Since high blood pressure (RestBP) is associated with heart disease, it is likely that people with lower MaxHR have slightly higher RestBP.

## HeatMap - correlation between multiple variables

```python
plt.figure(figsize=(10, 6))
correlation_matrix = df[['Age', 'RestBP', 'Chol', 'MaxHR', 'Oldpeak']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Heatmap: Correlation Matrix of Numerical Features')
plt.show()
```
